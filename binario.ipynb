```python
def float2bin(N):

  """
    Dado N un número decimal, devuelve su representación en binario
    

    Parámetros:
    * N: número decimal
    

    Valor de retorno 
    * representación binario
  """

  Nb=abs(N)
  integ=int(Nb)
  dec=Nb-float(int(Nb))
  a=""
  while dec != 0.0:
    dec=dec*2
    if dec < 1:
      a=a+"0"
    if dec >= 1:
      a=a+"1"
      dec=dec-1
    
  if (N<0):
    return "-" + bin(integ).replace("0b","")+"."+ a.replace("0b","")
  if (N>0):
    return  bin(integ).replace("0b","")+"."+ a.replace("0b","")
  return "0"

def bin2float(B):

  """
    Dado B un número binario, devuelve su representación en decimal
    

    Parámetros:
    * B: número binario
    
    
    Valor de retorno 
    * representación decimal
  """

  B=B.replace("0b","")
  entera=""
  decimal=""
  flot=0
  exp=1
  for i in B:
    if i==".":
      break
    entera=entera + i
    B=B[1:] 
  for i in B:
    decimal=decimal + i  
  for i in entera:
    if i=="1":
      flot=flot + 2**(len(entera)-exp)
    exp=exp+1
  exp=0
  for i in decimal:
    if i=="1":
      flot= flot + 1/2**(exp)
    exp=exp+1  
  return flot   
```

```python
   def binary_representation(f, n_mantisa = 52, n_exp = 11, complete_zeros = False ):
      """
        Dado f un número decimal, devuelve su representación en 
        signo, mantisa y exponente
    

        Parámetros:
        * f: número decimal
        * n_mantisa: número de dígitos de la mantisa del número
        * n_exp: número de dígitos del exponente del número
    
        Valor de retorno 
        * signo | mantisa | exponente
      """
      fl = float(abs(f)); 
      mantisa = float2bin(fl)
      exp=0
      if (fl >= 1):
        for i in mantisa:
          if i==".":
            mantisa=mantisa.replace(".","")
            break
          exp=exp+1
        mantisa= mantisa[1:]
        if complete_zeros:
          if (n_mantisa-len(mantisa))>0:
            for i in range(n_mantisa-len(mantisa)):
              mantisa=mantisa+ "0"
            else:
              mantisa=mantisa[:(n_mantisa+1)]
        exp=exp + int(math.pow(2, n_exp-1)-1)-1

        if f<0 :
           return "1"+"|"+mantisa +"|"+bin(exp).replace("0b","")     
        
        return "0"+"|"+mantisa +"|"+ bin(exp).replace("0b","")  
      
      if  fl<1 :
        exp=1
        mantisa=mantisa[2:]
        for i in mantisa:
          if i=="1":
            mantisa=mantisa[1:]
            break
          mantisa=mantisa[1:]
          exp=exp+1
        if complete_zeros: 
          if (n_mantisa-len(mantisa))>0:
            for i in range(n_mantisa-len(mantisa)):
              mantisa=mantisa+ "0"
            else:
              mantisa=mantisa[:(n_mantisa+1)]
        exp=exp+math.pow(2, n_exp-1)-1
        
        exp = int(exp)
        if f<0 :
           return "1"+"|"+mantisa +"|"+bin(exp).replace("0b","")    
        
        return "0"+"|"+mantisa +"|"+ bin(exp).replace("0b","")
```
